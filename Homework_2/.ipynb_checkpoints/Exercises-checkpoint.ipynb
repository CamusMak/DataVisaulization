Ex 4: Write a Python program to plot two or more lines on same plot with suitable legends of each line

```python
plt.figure(figsize=(15,10))
lines = plt.plot(df.Date, (df.Open, df.High, df.Low, df.Close,df.Close), marker = 'o')
plt.setp(lines[0], label = 'Open', color = 'orange')
plt.setp(lines[1], label = 'High', color = 'green')
plt.setp(lines[2], label = 'Low',  color = 'blue')
plt.setp(lines[3], label = 'Close',color = 'red')

plt.legend()
plt.show()

```

Ex 5: Write a Python program to plot two or more lines with legends, different widths and colors. 

```python


plt.figure(figsize=(15,10))

lines = plt.plot(df.Date, (df.Open, df.High, df.Low, df.Close,df.Close), marker = 'o',linewidth=12, markersize=32)
plt.setp(lines[0], label = 'Open', linewidth=12,color = 'orange')
plt.setp(lines[1], label = 'High', linewidth=12,color = 'green')
plt.setp(lines[2], label = 'Low',  linewidth=12,color = 'blue')
plt.setp(lines[3], label = 'Close',linewidth=12,color = 'red')

plt.legend()
plt.show()
```

Ex 6: Write a Python program to plot two or more lines with different styles.

```python

plt.figure(figsize=(15,10))

lines = plt.plot(df.Date, (df.Open, df.High, df.Low, df.Close,df.Close), marker = 'o',linewidth=5, markersize=32)
plt.setp(lines[0], label = 'Open', linewidth=5, linestyle="-." ,color = 'orange')
plt.setp(lines[1], label = 'High', linewidth=5, linestyle= "--",color = 'green')
plt.setp(lines[2], label = 'Low',  linewidth=5, linestyle= "-.",color = 'blue')
plt.setp(lines[3], label = 'Close',linewidth=5, linestyle= ":",color = 'red')

plt.legend()
plt.show()
```

Ex 7: Write a Python program to plot two or more lines and set the line markers. 

```python
plt.figure(figsize=(15,10))

x = np.array([0,4,5,6,7])
y = np.array([2,6,3,6,3])

plt.plot(x,y, color = 'red', 
         linestyle="dashdot",
         linewidth=5, 
         marker='o',
         markersize=45,
         markerfacecolor='orange',
         markeredgecolor='blue')

plt.title("Display marker", size=50)
plt.xlabel("x-axis", size=45)
plt.ylabel("y-label", size=45)

plt.show()
```

Ex 9: Write a Python program to plot several lines with different format styles in one command using arrays. 

```python
plt.figure(figsize=(15,7))

x = np.linspace(1,4,30)
y = x
z = x**2
m = x**3.5

plt.scatter(x,y, marker="_")

plt.scatter(x,z, marker='s', linewidths=7)

plt.scatter(x,m, marker="^", linewidths=7)


plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4,5]
x_label = ['Age group 1', 'Age group 2', 'Age group 3', 'Age Group 4', 'Age Group 5']
y = [10,20,30,50,80]
z = [1,7,4,5,8]

plt.plot(x,y, label = 'First chart')
plt.xlabel('X label')
plt.ylabel('Y label', fontsize=14, color='green')
# plt.xlim(0,6)
# plt.ylim(60,200)
# plt.axis([0, 6, 60, 200])
# plt.yscale('log')
plt.xticks(x, x_label, rotation = 20)
plt.yticks(y, rotation = 20)
plt.title('y vs x', fontsize=20, color='green')
plt.grid(True)
plt.text(3.5, 45, 'text on chart')
plt.text(3.5, 4.5, r'$mu=alpha + 2$')
plt.annotate('first annotation', xy=(2, 20), xytext=(2.2, 22),
    arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
# plt.show()
```
